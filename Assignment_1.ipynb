```python
from plotnine import *
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("~/Desktop/ChapmanSpring20/CPSC392/Datasets/iris.csv")
data.head()
```

Now that the data is imported we can use a plot to visualize the data. Let's plot sepal_length against petal_length and color the points by species.

```python
ggplot(data, aes(x = 'petal_length', y = 'sepal_length', color = 'species')) + geom_point() +geom_smooth() + labs(title = "Iris Length (sepal vs. petal)")
```

From this plot we learn a few things about the dataset

### Species Size
- The general size if each species varies. We see that *setosa* is the smaller of the species followed by *versicolor*, and *virginica* is, on average the larger of the three species in the dataset.

### Correlation
- The data on *versicolor* and *virginica* exhibits a strong positive linear relationship, that is it appears that as petal length increases, sepal length increases as well
- The data on *setosa* does not appear to show a relationship between petal and sepal length and appears to be more random than that of the other species.
